```python
x=['Sr_No', 'SeriousDlqin2yrs','DebtRatio_new', 'DebtRatio_flag',
       'MonthlyIncome_new', 'MonthlyIncome_flag', 'monthly_utilization_new',
       'age_new', 'NumberOfTime30-59DaysPastDueNotWorse_new',
       'NumberOfOpenCreditLinesAndLoans_new', 'NumberOfTimes90DaysLate_new',
       'NumberRealEstateLoansOrLines_new',
       'NumberOfTime60-89DaysPastDueNotWorse_new', 'NumberOfDependents_new']

loan_cleaned=loan[x]
loan_cleaned.head()
```

#Model Building

```python
# train and test data in logistics regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import pandas as pd
```

```python
# create x and y data

X=loan_cleaned.drop(['Sr_No','SeriousDlqin2yrs'],axis=1)
y=loan_cleaned['SeriousDlqin2yrs']

```

```python
# create train and test of x and y

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=123)
```

```python
#Print the shape
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Building logistics regression

log_reg=LogisticRegression()
log_reg.fit(X_train,y_train)
```

# Model Validation

```python
# Confussion Matrix and Accuracy on Train and Test data

y_pred_train=log_reg.predict(X_train)
y_pred_test=log_reg.predict(X_test)
```

```python
# confussion Matrix of Train data

confusion_matrix(y_train,y_pred_train)
```

```python
# confussion Matrix on test data

confusion_matrix(y_test,y_pred_test)
```

```python
# Accuracy on train and test data

print(accuracy_score(y_train,y_pred_train))
print(accuracy_score(y_test,y_pred_test))
```

# Multicollinearity

```python
import statsmodels.formula.api as sm

def vif_cal(input_data, dependent_col):
    x_vars=input_data.drop([dependent_col], axis=1)
    xvar_names=x_vars.columns
    for i in range(0,xvar_names.shape[0]):
        y=x_vars[xvar_names[i]]
        x=x_vars[xvar_names.drop(xvar_names[i])]
        rsq=sm.ols(formula="y~x", data=x_vars).fit().rsquared
        vif=round(1/(1-rsq),2)
        print (xvar_names[i], " VIF = " , vif)
```

```python
vif_cal(loan_cleaned,"SeriousDlqin2yrs")
```

```python
# Finding P value

#P_value of all the features
import statsmodels.api as sm
X_train_lm=sm.add_constant(X_train)
logistic_reg=sm.OLS(y_train,X_train_lm).fit()
print(logistic_reg.summary())
```
